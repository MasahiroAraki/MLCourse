# 3. 識別 ー概念学習ー

課題: wineデータで決定木の作成、表示を行ってください。また、過学習の状況および過学習への対処をコーディングで確認してください。

```
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import cross_val_score
```

```
wine = load_wine()
X = wine.data
y = wine.target
```

まず、全学習データで決定木を作成し、表示します。

```
clf = DecisionTreeClassifier()
clf.fit(X, y)
```

```
plot_tree(clf, filled=True, feature_names=wine.feature_names, class_names=wine.target_names)
plt.show()
```

次に、交差確認法で性能評価を行います。

```
scores = cross_val_score(clf, X, y, cv=10)
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std()*2))
```

木の深さを変えてみます。

```
clf2 = DecisionTreeClassifier(max_depth=2)
scores = cross_val_score(clf2, X, y, cv=10)
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std()*2))
```

```
clf3 = DecisionTreeClassifier(max_depth=8)
scores = cross_val_score(clf3, X, y, cv=10)
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std()*2))
```

学習データとの性能の差を見ます。

```
clf.fit(X, y)
clf.score(X,y)
```

```
clf2.fit(X, y)
clf2.score(X,y)
```

```
clf3.fit(X, y)
clf3.score(X,y)
```

学習データを用いた評価と交差確認法で順位が入れ替わっていれば過学習の確認ができますが、このデータでは入れ替わっていないので、デフォルトのパラメータでも過学習は起きていないと予測されます。
