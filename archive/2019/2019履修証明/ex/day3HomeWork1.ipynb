アッシェンフェルターのワイン方程式を導く元となったデータ http://www.liquidasset.com/winedata.html からCSVファイルを作成し、回帰式を導け。ただし回帰式の学習には1980年までのデータを用いること。*斜体テキスト*

webページでデータの該当範囲を選択・コピーし、表計算ソフトにペーストしてCSVファイルを作成します。ターゲットが欠損しているデータは削除します。また、通し番号および年を表す列(OBS, VINT)も削除しておきます。

## 準備

必要なライブラリ等を読み込みます。

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

## データの読み込み

wine.csvをランタイムにアップロードしてPandasのデータフレームとして読み込みます。

```python
wine = pd.read_csv('wine.csv')
```

列名を確認します。

```python
wine.columns
```

ターゲットをy、特徴をXに格納します。

```python
y = wine[['LPRICE2']]
X = wine[['WRAIN', 'DEGREES', 'HRAIN', 'TIME_SV']]
```

線形回帰を実行します。

```python
lr = LinearRegression()
lr.fit(X, y)
```

係数と切片の値を表示します。

```python
for f, w in zip(X.columns, lr.coef_[0]) :
    print(f, w)
print('INTERCEPT ', lr.intercept_[0])
```
