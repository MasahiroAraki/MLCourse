## 準備

使用するパッケージを読み込みます。

```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.preprocessing import scale
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
```

## データの読み込み

irisデータはscikit-learnに組み込みで用意されています。  
http://scikit-learn.org/stable/datasets/index.html  
load_iris()メソッドの戻り値はBunchオブジェクトで、キーdataに特徴ベクトル、キーtargetに正解ラベルが格納されています。

```python
iris = load_iris()
print(iris.DESCR)
```

```python
iris.data
```

```python
iris.target
```

## 標準化

特徴ベクトルを標準化したものをXに格納します。

```python
X = scale(iris.data)
X
```

各次元の平均・分散を確認します。  

引数axisの値は0が列方向、1が行方向です。省略すると全要素に対する計算になります。

```python
X.mean(axis=0)
```

```python
X.var(axis=0)
```

正解ラベルをyに格納します。

```python
y = iris.target
```

## 学習

1-NN法による識別を行うインスタンスを作成します

```python
clf = KNeighborsClassifier(n_neighbors=1)
clf.fit(X,y)
```

交差確認法を実行し、結果を格納します。

```python
score = cross_val_score(clf, X, y, cv=10)
score
```

## 結果の表示

識別率の平均と標準偏差を表示します。

```python
print("{0:4.2f} +/- {1:4.2f} %".format(score.mean() * 100, score.std() * 100))
```
