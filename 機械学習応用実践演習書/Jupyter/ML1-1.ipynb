数値計算ライブラリNumpyとグラフ描画ライブラリMatplotlibの読み込み

```python
import numpy as np
import matplotlib.pyplot as plt
```

CSVファイルを読み込み、Xに特徴ベクトル、yに教師データを格納

```python
data = np.loadtxt("jissen8-1.csv",delimiter=",")
# 分散x(3列目)と分散y(4列目)をXに格納
X=data[:,2:4]
# 教師データは整数としてyに格納
y=data[:,8].astype(np.int)
```

Matplotlibを使ったグラフの表示

```python
# ヒストグラム
fig1 = plt.figure()
ax1 = fig1.add_subplot(111)
ax1.set_xlabel("Variance X")
ax1.set_ylabel("Freq")
ax1.hist(X[:,0])
```

```python
# 散布図
fig2 = plt.figure()
ax2 = fig2.add_subplot(111)
ax2.set_xlabel("Variance X")
ax2.set_ylabel("Variance Y")
ax2.plot(X[:,0], X[:,1],".")
```

```python
#クラス毎に色分けしてプロット
fig3 = plt.figure()
ax3 = fig3.add_subplot(111)
ax3.set_xlabel("Variance X")
ax3.set_ylabel("Variance Y")
colors = (i + j for j in 'o<.' for i in 'bgrcmyk')
labels = 'zero one two three four five six seven eight nine'.split()
for i, l, c  in zip(range(10), labels, colors):
    start, stop = i * 10, (i + 1) * 10
    ax3.plot(X[start:stop,0], X[start:stop,1], c, label=l)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```
