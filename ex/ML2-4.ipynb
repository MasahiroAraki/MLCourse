# 実践演習2-4

ライブラリの読み込み

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.decomposition import PCA
from sklearn import preprocessing
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import confusion_matrix
```

### 学習データ (breast cancer) の読み込みと概要の表示

```python
bc = load_breast_cancer()
print(bc.DESCR)
X = bc.data
y = bc.target
```

### 前処理（主成分分析）

```python
pca = PCA()
X2 = pca.fit_transform(X)
```

結果を散布図として表示

```python
plt.plot(X2[y==0,0], X2[y==0,1],"ro")
plt.plot(X2[y==1,0], X2[y==1,1],"g.")
```

累積寄与率の表示

```python
ev_ratio = pca.explained_variance_ratio_
ev_ratio = np.hstack([0,ev_ratio.cumsum()])
plt.plot(ev_ratio)
```

### 前処理 (標準化)

```python
X_scaled = preprocessing.scale(X)
```

### 学習（k-NN法で識別を行う識別器の作成）

```python
clf = KNeighborsClassifier(n_neighbors=1)
print(clf)
```

学習（交差確認法による性能評価）

```python
scores = cross_val_score(clf, X_scaled, y, cv=10)
```

正解率の平均と95%信頼区間の表示

```python
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std()*2))
```

学習（混同行列の表示）

```python
y_pred = cross_val_predict(clf, X_scaled, y, cv=10)
print(confusion_matrix(y, y_pred))
```

kの値を変えて評価

```python
clf2 = KNeighborsClassifier(n_neighbors=3)
scores = cross_val_score(clf2, X_scaled, y, cv=10)
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std()*2))
```

```python
clf3 = KNeighborsClassifier(n_neighbors=5)
scores = cross_val_score(clf3, X_scaled, y, cv=10)
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std()*2))
```

```python
clf4 = KNeighborsClassifier(n_neighbors=7)
scores = cross_val_score(clf4, X_scaled, y, cv=10)
print("Accuracy: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std()*2))
```
