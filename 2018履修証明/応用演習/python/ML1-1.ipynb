CSVファイルを読み込み、様々なグラフとして表示します。

数値計算ライブラリNumpyとグラフ描画ライブラリMatplotlibの読み込み

```python
import numpy as np
import matplotlib.pyplot as plt
```

CSVファイルを読み込み、Xに特徴ベクトル、yに教師データを格納

```python
data = np.loadtxt("numbers.csv",delimiter=",")
# 分散x(3列目)と分散y(4列目)をXに格納
X=data[:,2:4]
# 教師データは整数としてyに格納
y=data[:,8].astype(np.int)
```

Matplotlibを使ったグラフの表示

```python
# ヒストグラム
plt.hist(X[:,0])
```

```python
# 散布図
plt.plot(X[:,0], X[:,1], "bo")
```

```python
#クラス毎に色分けしてプロット
colors = (i + j for j in 'o<.' for i in 'bgrcmyk')
labels = 'zero one two three four five six seven eight nine'.split()
for i, l, c  in zip(range(10), labels, colors):
    start, stop = i * 10, (i + 1) * 10
    plt.plot(X[start:stop,0], X[start:stop,1], c, label=l)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

凡例を省略してもよければ、もっと簡単な方法があります。

```python
plt.set_cmap(plt.cm.brg)
plt.scatter(X[:, 0], X[:, 1], c=y)
```
